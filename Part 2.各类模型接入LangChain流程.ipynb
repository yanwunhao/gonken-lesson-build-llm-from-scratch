```python
import os
from dotenv import load_dotenv 
load_dotenv(override=True)

DeepSeek_API_KEY = os.getenv("DEEPSEEK_API_KEY")
# print(DeepSeek_API_KEY)  # å¯ä»¥é€šè¿‡æ‰“å°æŸ¥çœ‹
```

&emsp;&emsp;æˆ‘ä»¬åœ¨å½“å‰çš„è¿è¡Œç¯å¢ƒä¸‹ä¸ä½¿ç”¨`LangChain`ï¼Œç›´æ¥ä½¿ç”¨`DeepSeek`çš„`API`è¿›è¡Œç½‘ç»œè¿é€šæ€§æµ‹è¯•ï¼Œæµ‹è¯•ä»£ç å¦‚ä¸‹ï¼š

```python
# ! pip install openai
```

```python
from openai import OpenAI

# åˆå§‹åŒ–DeepSeekçš„APIå®¢æˆ·ç«¯
client = OpenAI(api_key=DeepSeek_API_KEY, base_url="https://api.deepseek.com")

# è°ƒç”¨DeepSeekçš„APIï¼Œç”Ÿæˆå›ç­”
response = client.chat.completions.create(
    model="deepseek-chat",
    messages=[
        {"role": "system", "content": "ä½ æ˜¯ä¹äºåŠ©äººçš„åŠ©æ‰‹ï¼Œè¯·æ ¹æ®ç”¨æˆ·çš„é—®é¢˜ç»™å‡ºå›ç­”"},
        {"role": "user", "content": "ä½ å¥½ï¼Œè¯·ä½ ä»‹ç»ä¸€ä¸‹ä½ è‡ªå·±ã€‚"},
    ],
)

# æ‰“å°æ¨¡å‹æœ€ç»ˆçš„å“åº”ç»“æœ
print(response.choices[0].message.content)
```
